```python
import requests
import json
import datetime
import logging
import os
import time
import zipfile

# Calculate yesterday
today = datetime.date.today()
yesterday = today - datetime.timedelta(days=1)

# If today is Monday (weekday() function returns 0 for Monday) then set yesterday to be last Friday
if today.weekday() == 0:
    yesterday = today - datetime.timedelta(days=3)

# Convert yeserday variable to string
yesterday = str(yesterday)

# Calculate start and end time for the given day
start = yesterday + 'T' + '00:00:00Z'
end = yesterday + 'T' + '23:59:59Z'

RingbaAuthApiUrl = "https://api.ringba.com/v2/token"
RigbaApiUrl = "https://api.ringba.com/v2/RA0909aa9eddae40c4bb48f7253e6e8742/calllogs/export/csv"

RingbaUsername = 'mike%40carida.com'
RingbaPassword = 'Rutgers-72'

# ---------------------------------------------

payload = 'grant_type=password&username=' + RingbaUsername + '&password=' + RingbaPassword
headers = {
  'content-type': 'application/x-www-form-urlencoded; charset=UTF-8'
}

response = requests.request("POST", RingbaAuthApiUrl, headers=headers, data=payload)

RingbaAuthToken = json.loads(response.text)['access_token']

# ---------------------------------------------

# Adding RingbaAuthToken to the log file adding in the current directory a new line and date and time
logging.basicConfig(filename='Dispos2.log', level=logging.DEBUG)
logging.info('\n' + str(datetime.datetime.now()) + ' ' + RingbaAuthToken)

# ---------------------------------------------

```

```python
# Calling Ringba API to order a CSV file and filename.

payload = json.dumps({
  "reportEnd": end,
  "reportStart": start,
  "orderByColumns": [
      {
          "column": "callDt",
          "direction": "desc"
      }
  ],
"valueColumns": [
    {
      "column": "callDt"
    },
    {
      "column": "publisherName"
    },
    {
      "column": "number"
    },
    ]
})
headers = {
  'Authorization': 'Bearer' + ' ' + RingbaAuthToken,
  'Content-Type': 'application/json'
}

response = requests.request("POST", RigbaApiUrl, headers=headers, data=payload)

# ---------------------------------------------

# Handling exceptions and writing them to the log file call the API and write it into CsVid variable
try:
    CsVid = json.loads(response.text)['id']
except:
    logging.info('\n' + str(datetime.datetime.now()) + ' ' + response.text)
    CsVid = 'Error'

# ---------------------------------------------

```

```python





url = "https://api.ringba.com/v2/RA0909aa9eddae40c4bb48f7253e6e8742/calllogs/export/" + CsVid
payload={}
headers = {
  'Authorization': 'Bearer' + ' ' + RingbaAuthToken
  }

# Running response with get, handling exceptions and writing them to the log file
try:
    response = requests.request("GET", url, headers=headers, data=payload)
    r = response.json()
    logging.info('\n' + str(datetime.datetime.now()) + ' ' + response.text)  
except:
    logging.info('\n' + str(datetime.datetime.now()) + 'Error:  ' + response.text)

# poll the API until the file is ready

while r['status'] != 'Ready':
    time.sleep(5)
    response = requests.request("GET", url, headers=headers, data=payload)
    r = response.json()
    logging.info('\n' + str(datetime.datetime.now()) + 'URL is not ready for some reason ' + response.text)
    break

print(r)

filename = url.split('/')[-1]

requestedURLName = requests.get(r['url'])


# extract the zip file in the current directory
with open(filename, 'wb') as f:
    f.write(requestedURLName.content)
    with zipfile.ZipFile(filename, 'r') as zip_ref:
        zip_ref.extractall(os.getcwd())
        
        logging.info('\n' + str(datetime.datetime.now()) + ' ' + 'Zip file extracted')
```
