```python
import requests
import datetime
import logging
import json


# Calculate yesterday
today = datetime.date.today()
yesterday = today - datetime.timedelta(days=1)

# If today is Monday, set yesterday to be last Friday
if today.weekday() == 0:
    yesterday = today - datetime.timedelta(days=3)

# Convert yesterday variable to string
yesterday = str(yesterday)

# Calculate start and end time for the given day
start = yesterday + ' ' + '00:00:00'
end = yesterday + ' ' + '23:59:59'
auth_token = "vu332i5d8jx6ppqyhmkq76sxs2e2w6fh"



# Set the desired parameters for filtering call log records
params = {
    'auth_token': auth_token,
    'start_time': start,
    'end_time': end,
    'limit': '500'
}


# Make the API request
url = 'https://api.convoso.com/v1/log/retrieve'
headers = {
    'Content-Type': 'application/json',
    'Authorization': 'Bearer' + auth_token
}



# making a post API request
response = requests.post(url, headers=headers, params=params)

data = response.json()

#formatting the json output
#print(json.dumps(data, indent=4, sort_keys=True))
lists = data['data']['results']
#phone_numbers = [item['phone_number'] for item in lists]
print(lists)

# check if lists arenot empty

if lists is not None: 
    # Extract the phone numbers and list IDs
    phone_numbers = [item['phone_number'] for item in lists]
    list_ids = [item['list_id'] for item in lists]
    # Print the results
    for phone_number, list_id in zip(phone_numbers, list_ids):
        # Extracting the phone numbers and list IDs into csv file. 
        with open('ConvosoListId_Phone.csv', 'a') as f:
            f.write('+1' + phone_number + ',' + list_id + '\n')
            #Adding sum of the lines to log file with date and time stamp
            logging.basicConfig(filename='dispos_next.log', level=logging.INFO)
            logging.info('Phone Number: %s, List ID: %s', phone_number, list_id)
        # print('---')  



    

        # print('Phone Number:', phone_number)
        # print('List ID:', list_id)
        # print('---')
else:
    print('No logs found for the specified time range.')
#else:
#print('Invalid response or empty data.')



#print headers and status code
# print(response.headers)
# print(response.status_code)
# print(response.url)

```
